```python
import numpy as np
import pandas as pd

print("===============================")
print("|Load Data for prediction")
print("===============================")

rewardDict = dict()
#Music | Speech
#For combating float point precision the rewards have been multiplyed with 10
rewardDict[(0,0)] = 0
rewardDict[(0,1)] = 0
rewardDict[(1,0)] = 10
rewardDict[(1,1)] = 2

estimated_profit = 0

indeces = range(15,30)
filename_music = "Predictions/{}.music.txt"
filename_speech = "Predictions/{}.speech.txt"
for index in indeces:
    print("|")
    print("|Load Data from Index: ",index)
    pred_music = np.genfromtxt(filename_music.format(index), delimiter='\n')
    pred_speech = np.genfromtxt(filename_speech.format(index), delimiter='\n')
    data = {'music': pred_music, 'speech': pred_speech}
    df = pd.DataFrame(data=data, dtype=np.int64)
    profit_one_round = np.sum([rewardDict[(x,y)] for x, y in zip(df.music, df.speech)])
    estimated_profit += profit_one_round
    #To undo the times 10 performed on the rewards
    print("|Estimated profit for this file: ", profit_one_round/10)
    print("|Estimated profit after this round: ", estimated_profit/10)
    print("===============================")
```
